```python
import pandas as pd
from dataprep.eda import plot, plot_correlation, plot_missing
from dataprep.clean import clean_headers, clean_df
from datacleaner import autoclean
import matplotlib.pyplot as plt


# Анализ пропусков
plot_missing(df2)

# Корреляционный анализ
plot_correlation(df2)

# Очистка заголовков (удаление пробелов, спецсимволов)
df2_clean = clean_headers(df2)

# Автоматическая очистка данных
df2_clean = clean_df(df2_clean, 
                    replace_missing={"method": "auto"},  # авто-замена (медиана/мода)
                    standardize="auto",                  # стандартизация текста/чисел
                    clean_text=True                      # очистка текстовых данных
                   )

df2_clean = autoclean(df2_clean)
df2_clean.head()
df2_clean.dtypes
```
